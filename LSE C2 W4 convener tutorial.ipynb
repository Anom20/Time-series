```python
# import libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
# import CSV file
weather_data = pd.read_csv("ny_hourly.csv")
weather_data.shape
```

```python
# sense-check data set
weather_data
```

```python
# read only four columns
weather_data = pd.read_csv("ny_hourly.csv", 
                           parse_dates=[["date", "TimeEST"]],
                           usecols=["date", "TimeEST", "TemperatureF", "Dew PointF", "Humidity"])
weather_data
```

```python
# include data where temperature >-100
weather_data_cleaned = weather_data[weather_data.TemperatureF > -100]
weather_data_cleaned
```

```python
# sort the data according to date
weather_data_sorted = weather_data_cleaned.sort_values("date_TimeEST")
weather_data_sorted
```

# 

```python
# data locator and formatters
# date on x-axis and temperature on y-axis
sns.lineplot(x="date_TimeEST", y="TemperatureF", 
             data=weather_data_sorted).set_title("Hourly Weather in New York")
```

```python
# date on x-axis and temperature on y-axis
sns.lineplot(x="date_TimeEST", y="Humidity", 
             data=weather_data_sorted).set_title("Hourly Weather in New York")
```

```python
# date on x-axis and temperature on y-axis
sns.lineplot(x="date_TimeEST", y="Dew PointF", 
             data=weather_data_sorted).set_title("Hourly Weather in New York")
```

```python
# format dates
fig, ax = plt.subplots()
fig = sns.lineplot(x="date_TimeEST", y="TemperatureF", data=weather_data_sorted).set_title("Hourly Weather in New York")
x_labels = weather_data_sorted["date_TimeEST"].dt.strftime("%Y-%m-%d")
ax.set_xticklabels(x_labels, rotation=45)
plt.show()
```
